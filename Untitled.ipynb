### visualisations

```python
sns.heatmap(df1.isna(), cmap='viridis', cbar=False)
plt.title("Heatmap des valeurs manquantes")
plt.show()
```

```python
counts = df3['prdtypecode'].value_counts()
counts.plot(kind='bar', color='skyblue')
plt.title("Répartition des types de produits")
plt.xlabel("prdtypecode")
plt.ylabel("Nombre d'occurrences")
```

```python
missing_description = df1_new ['description'].isna().sum()
total = len(df1_new )
present_description = total - missing_description

# Visualisation avec un pie chart
labels = ['Présente', 'Manquante']
sizes = [present_description, missing_description]
colors = ['lightgreen', 'lightcoral']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title("Proportion de descriptions manquantes")
plt.axis('equal')  # Assure que le pie chart est circulaire
plt.show()
```

```python
plt.scatter(df1_new['productid'],df1_new['imageid'], color='blue')
plt.title("Relation entre ProductID et ImageID")
plt.xlabel("ProductID")
plt.ylabel("ImageID")
plt.show()
```

```python
plt.hist(df1_new['productid'], bins=5, color='skyblue', edgecolor='black')
plt.title("Distribution des ProductID")
plt.xlabel("ProductID")
plt.ylabel("Fréquence")
plt.show()
```

```python
plt.boxplot(df1_new['productid'])
plt.title("Box Plot des ProductID")
plt.ylabel("ProductID")
plt.show()
```
